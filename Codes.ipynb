```python
import pandas as pd
import requests

base_url = "https://statsapi.web.nhl.com/api/v1/people/{}/stats?stats=statsSingleSeason&season={}"
pim = []
power_play_goal = []
over_time_gaol = []
plus_mins = []
points = []

for index, row in winning_df.iterrows():
    player_id = row["Player ID"]
    year = row["Season"]
    url = base_url.format(player_id, year)
    response = requests.get(url)
    data = response.json()
    
    # Check if 'stats' key exists in the response
    if "stats" in data:
        stats = data["stats"]
        if stats:
            splits = stats[0].get("splits", [])
            if splits:
                pim_data = splits[0]["stat"].get("pim", None)
                ppg_data = splits[0]["stat"].get("powerPlayGoals", None)
                otg_data = splits[0]["stat"].get("overTimeGoals", None)
                pm_data = splits[0]["stat"].get("plusMinus", None)
                points_data = splits[0]["stat"].get("points", None)
            else:
                pim_data = None
                ppg_data = None
                otg_data = None
                pm_data = None
                points_data = None
        else:
            pim_data = None
            ppg_data = None
            otg_data = None
            pm_data = None
            points_data = None
    else:
        pim_data = None
        ppg_data = None
        otg_data = None
        pm_data = None
        points_data = None 
    pim.append(pim_data)
    power_play_goal.append(ppg_data)
    over_time_gaol.append(otg_data)
    plus_mins.append(pm_data)
    points.append(points_data)      
winning_df['PIM'] = pim    
winning_df['PPG'] = power_play_goal
winning_df['OTG'] = over_time_gaol
winning_df['PM'] = plus_mins
winning_df['Pts'] = points         
```
