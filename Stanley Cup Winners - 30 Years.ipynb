```python
import pandas as pd
import datetime
from random import randint
from time import sleep
import os
```

```python
# Get game dates of games played

url = "https://en.wikipedia.org/wiki/List_of_Stanley_Cup_champions"

stanleycup_shamps = pd.read_html(url)

stanleycup_byyear_df = stanleycup_shamps[2]
stanleycup_byyear_df.head()
```

```python
stanleycup_cleaned_df = stanleycup_byyear_df.loc[stanleycup_byyear_df['Year'].between(1991,2022)]
stanleycup_cleaned_df = stanleycup_cleaned_df[["Year", "Winning team"]]
stanleycup_cleaned_df["Winning team"] = stanleycup_cleaned_df["Winning team"].str.split('(').str[0]

stanleycup_cleaned_df.loc[stanleycup_cleaned_df["Winning team"]== 'cancelled']
stanleycup_cleaned_df.drop(stanleycup_cleaned_df[(stanleycup_cleaned_df['Year'] == 2005)].index, inplace=True)

stanleycup_cleaned_df
```
